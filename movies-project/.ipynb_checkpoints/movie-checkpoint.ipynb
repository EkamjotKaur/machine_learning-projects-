```python
import pandas as pd
movie_data=pd.read_csv('movie_reviews.csv',parse_dates=True)
movie_data.head(5)
```

```python
columns=['title','actors','directors','hitFlop']
data=movie_data[columns]
data.head(5)
```

```python
actors_data=pd.read_csv('BollywoodActorRanking.csv')
directors_data=pd.read_csv('BollywoodDirectorRanking.csv')
```

```python
actors_data.head(5)
```

```python
col1=['actorName','normalizedRating']
data1=actors_data[col1]
data1.set_index('actorName',inplace=True)
#data1.rename(columns={'normalizedRating':'ActorsRating'},inplace=True)
data1.head(5)
```

```python
directors_data.head(5)
```

```python
col2=['directorName','normalizedRating']
data2=directors_data[col2]
data2.set_index('directorName',inplace=True)
data2.head(5)
```

```python
data.head()
```

# removing nan values

```python
import numpy as np
from numpy import math
print(data['actors'].value_counts(dropna=False))
def replacenanvalues(actor_name):
    if isinstance(actor_name,float) and math.isnan(actor_name):
        return str(0)
    else:
        return str(actor_name)
data['actors']=data['actors'].apply(replacenanvalues)
data['actors'].value_counts(dropna=False) 
```

# giving average of the ranks of the cast in movie

```python
def comparison(actor_name):
    lst=[]
    s=0
    if isinstance(actor_name,float) and math.isnan(actor_name):
        return 0
    else:
        lst=actor_name.split(' | ')
    for i in lst:
        if i in data1.index:
            s=s+data1.normalizedRating.loc[i]
        else:
            s=s+0
    return s/len(lst)
data['rate_actor']=data['actors'].apply(comparison)
data
```

```python
print(data['directors'].value_counts(dropna=False))
data['directors']=data['directors'].apply(replacenanvalues)
print(data['directors'].value_counts(dropna=False))
```

# giving average value to directors of movie

```python
def compare(director_name):
    lst=[]
    s=0
    if isinstance(director_name,float) and math.isnan(director_name):
        return 0
    elif director_name not in data2.index:
        return 0
    else:
        if '|' in director_name:
            lst=director_name.split(' | ')
        for i in lst:
            if i in data2.index:
                s=s+data2.normalizedRating.loc[i]
            else:
                s=s+0
        else:
            s=s+data2.normalizedRating.loc[director_name]
        return s

data['rate_director']=data['directors'].apply(compare)
data.head(3)
```

```python
count=0#counting no of actors with zero rate
for i in data['rate_actor']:
    if i==float(0):
        count=count+1
print(count)
```

```python
count=0#counting no of directors with zero rate
for i in data['rate_director']:
    if i==float(0):
        count=count+1
print(count)
```

```python
data
```

# changing hitflop ratings:
- 1.flop
- 2.average
- 3.hit
- 4.superhit
- 5.blockbuster

```python
def changerating(hitFlop):
        if hitFlop==int(2) or hitFlop==int(1):
            return int(1)
        elif hitFlop==int(4) or hitFlop==int(3):
            return int(2)
        elif hitFlop==int(5) or hitFlop==int(6):
            return int(3)
        elif hitFlop==int(7):
            return int(4)
        elif hitFlop==int(8) or hitFlop==int(9):
            return int(5)
data['hitFlop']=data['hitFlop'].apply(changerating)
data['hitFlop'].value_counts(dropna=False)
```

```python
data.head(5)
```

# applying various models to data

- using train,test split

```python
from sklearn.model_selection import train_test_split#splitting train/test data
col1=['rate_actor','rate_director']
X=data[col1]
Y=data['hitFlop']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=4)
```

```python
from sklearn import metrics#apply kNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,Y_train)
y_pred=knn.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB#applying GaussianNB
nb=GaussianNB()
nb.fit(X_train,Y_train)
y_pred=nb.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn.linear_model import SGDClassifier#applying sgdclassifier(generally for large data)
sgd=SGDClassifier(loss='modified_huber',shuffle=True,random_state=5)
sgd.fit(X_train,Y_train)
y_pred=sgd.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier#applying decisiontreeclassifier
d_tree=DecisionTreeClassifier(random_state=5)
d_tree.fit(X_train,Y_train)
y_pred=d_tree.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier#applying randomforestclassifier(reduces overfitting)
rfm=RandomForestClassifier(n_estimators=12)
rfm.fit(X_train,Y_train)
y_pred=rfm.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn.svm import SVC#applying supportvector classifier
sv=SVC(kernel='linear',C=1,random_state=5)
sv.fit(X_train,Y_train)
y_pred=sv.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression#applying logistic regression
lg=LogisticRegression(solver='liblinear',multi_class='auto')
lg.fit(X_train,Y_train)
y_pred=lg.predict(X_test)
print(metrics.accuracy_score(Y_test,y_pred))
```

- using cross val score

```python
from sklearn.model_selection import cross_val_score
k_range=range(1,31)
k_scores=[]
for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn,X,Y,cv=4,scoring='accuracy')
    k_scores.append(scores.mean())
print(k_scores)
print(max(k_scores))
print(k_scores.index(max(k_scores)))
```

```python
#applying kNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
scores=cross_val_score(knn,X,Y,cv=10,scoring='accuracy',error_score='raise')
print(scores)
```

```python
dst=DecisionTreeClassifier(random_state=5)
print(cross_val_score(dst,X,Y,cv=10,scoring='accuracy').mean())
```

```python
rft=RandomForestClassifier(n_estimators=12)
print(cross_val_score(rft,X,Y,cv=10,scoring='accuracy').mean())
```

```python
def call_me():
    s2=entry1.get()
    s3=entry2.get()
    s4=entry3.get()
    lst=[]
    lst=s3.split(',')
    lst=[i.title() for i in lst]
    lst1=[]
    flag=0
    for actor_name in lst:
        if actor_name in data1.index:
            flag=1
            lst1.append(data1.normalizedRating[actor_name])
        else:
            continue
    if(flag==0):
        lst1.append(float(0))
    s=0
    for n in lst1:
        s=s+n
    av_rate_actors=s/len(lst1)
    lst2=[]
    lst2=s4.split(',')
    lst2=[i.title for i in lst2]
    for dir_name in lst:
        if dir_name in data2.index:
            flag=1
            lst1.append(data2.normalizedRating[dir_name])
        else:
            continue
    if(flag==0):
        lst1.append(float(0))
    s=0
    for n in lst1:
        s=s+n
    av_rate_directors=s/len(lst1)
    y_pred=knn.predict([[av_rate_actors,av_rate_directors]])
    if y_pred==1:
        s1.set('1:FLOP')
    elif y_pred==2:
        s1.set('2:AVERAGE')
    elif y_pred==3:
        s1.set('3:HIT')
    elif y_pred==4:
        s1.set('4:SUPER HIT')
    elif y_pred==5:
        s1.set('5:BLOCKBUSTER')
```
